# Using Pandas for simple analysis

```python
import pandas as pd
```

```python
# The following is current sales results for Allegheny County Sherif's auctions:
# linked from https://catalog.data.gov/dataset/allegheny-county-sheriff-sales

# It is possible to read in data from a url or a local file

df = pd.read_csv("https://data.wprdc.org/datastore/dump/adffb94b-6175-4612-8628-44e7c9baa620")
```

## Selecting data

```python
# Get values for a specified column - in Pandas this returns a "Series"

df['SaleDate']
```

```python
# Get a unique list of values for a specified column along with their corresponding value counts

df['SaleStatus'].value_counts()
```

```python
# Get a list of columns showing if there is missing values?
# Replace "any" with "all" to determine which columns have no values

df.isna().any()
```

## Querying our DataFrame

```python
# return dataframe with only "Mortgage Foreclosure"

# SQL example:
# SELECT * FROM df WHERE SaleType = "Mortgage Foreclosure";

df[df['SaleType'] == "Mortgage Foreclosure"]
```

```python
# Get the highest cost for POSTPONED sale

# SQL example:
# SELECT MAX(CostsTaxes) FROM df WHERE SalesStatus = "POSTPONED";

df[df['SaleStatus'] == "POSTPONED"]['CostsTaxes'].max()
```

```python
# Get the entire row for the lowest cost for a POSTPONED sale
# Note we are using index "idxmin" and location "loc" to return the entire row

# SQL example:
# SELECT * FROM df where CostsTaxes = (Select MIN(CostsTaxes) from df);

df.loc[df[df['SaleStatus'] == "POSTPONED"]['CostsTaxes'].idxmin()]
```

## Merging DataFrames

```python
df_sales = pd.read_csv('../data/final_sale.csv')
```

```python
df_sales
```

```python
# Merge the two dataframes returning common rows only (inner join)
df_cost_sale = df_sales.merge(df, how='inner', left_on='SaleNumber', right_on='DocketNumber')
df_cost_sale
```

## Applying functions to DataFrames

```python
def double(x):
    return x*2

df_cost_sale['SalePrice'].apply(double)
```

```python
# Example of using a lambda function to split text

df_cost_sale['DocketNumber'].apply(lambda x: x.split('-'))
```

```python
# Create a new column and store the results of our calculations

df_cost_sale['Recovered'] = df_cost_sale.apply(lambda x: x.SalePrice - x.CostsTaxes, axis=1)
```

```python
df_cost_sale
```

## Plotting Example

```python
import matplotlib.pyplot as plt
%matplotlib inline

df_cost_sale[['SalePrice', 'CostsTaxes', 'Recovered']].plot(kind='bar')
```

```python
# For readability, add title and labels to our plot

ax = df_cost_sale[['SalePrice', 'CostsTaxes', 'Recovered']].plot(kind='bar', title='Jan. 6th, 2020 - Sherif Auction')
ax.set_ylabel('Dollars $')
ax.set_xlabel('Dockets')
ax.set_xticklabels(df_cost_sale['DocketNumber'])
ax
```

## Timeseries Example
This section is heavily based on the [Pandas Time Series Documentation](https://pandas.pydata.org/pandas-docs/stable/getting_started/10min.html#time-series)

```python
# numpy is used to generate random data for this example

import numpy as np
```

```python
rng = pd.date_range('1/1/2012', periods=100, freq='Min')
rng
```

```python
ts = pd.Series(np.random.randint(0, 500, len(rng)), index=rng)
ts
```

```python
# Resample our 1 minute timeseries to 10 minute increments using "sum" to aggregate.

ts.resample('10Min').sum()
```
